```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("../data/raw/WineQT.csv")

# Remove Id column
df = df.drop("Id", axis=1)
```

```python
# Set plotting style
sns.set_style("whitegrid")

# Increase default figure size
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
# Plot histograms for all numerical features
df.hist(figsize=(16, 12), bins=20)

# Adjust spacing
plt.tight_layout()

# Display the plots
plt.show()
```

```python
# Count the occurrences of each wine quality score
plt.figure(figsize=(8, 5))

sns.countplot(
    x="quality",
    data=df,
    hue="quality",      # Required in newer Seaborn versions
    palette="viridis",
    legend=False
)

plt.title("Distribution of Wine Quality")
plt.xlabel("Wine Quality")
plt.ylabel("Number of Wines")

plt.show()
```

```python
# Create box plots for all numerical features
plt.figure(figsize=(15, 10))

sns.boxplot(data=df)

plt.title("Box Plot of All Features")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.heatmap(
    df.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x="quality", y="alcohol", data=df)

plt.title("Alcohol vs Wine Quality")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x="quality", y="volatile acidity", data=df)

plt.title("Volatile Acidity vs Quality")
plt.show()
```

```python
sns.pairplot(
    df[
        [
            "alcohol",
            "volatile acidity",
            "citric acid",
            "quality",
        ]
    ],
    hue="quality"
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.violinplot(x="quality", y="alcohol", data=df)

plt.title("Alcohol Distribution by Quality")
plt.show()
```
